### ARIMA

В основе моделей класса ARIMA лежит понятие автокорреляции и плавающего среднего.

<img width = '400px' src="images/arma.png">


$ p(AR(p)): y_{t} - \text {линейная комбинация p предыдущих значений ряда и шумовой компоненты}.$

$ q(MA(q)): y_{t} - \text {линейная комбинация q последних значений шумовой компоненты}.$

ARIMA - линейная модель множественной регрессии, в которой в качестве объяснящих переменных выступают прошлые значения самой зависимой переменной, а в качестве регрессионного остатка - скользящие средние из элементов белого шума.

Для корректного использование модели ARIMA исходный ряд должен быть стационарен.


```python
def initial_plot(df):
    
    initial_plot = plt.figure(num=1, figsize=(12, 6))
    plt.plot(df["temperatureMean"], color="navy", linewidth=0.5, figure=initial_plot)
    plt.title("Mean Temperature from 2010-2018", figure=initial_plot)
    plt.xlabel("Time", figure=initial_plot)
    plt.ylabel("Mean Temperature (˚C)", figure=initial_plot)
    plt.savefig("plots/initial_plot.png", dpi=200, figure=initial_plot)
```

```python
initial_plot(df)
```

```python
def stationarity_check(df, column):
    
    def adf_check(time_series):
        
        result = adfuller(time_series)
        print()
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print("Augmented Dickey-Fuller Unit Root Test:")
        labels = [
            "ADF Test Statistic",
            "p-value",
            "#Lags Used",
            "Number of Observations Used",
        ]

        for value, label in zip(result, labels):
            print(label + " : " + str(value))

        if result[1] <= 0.005:
            print(
                "Нет единичных корней, ряд стационарен"
            )
        else:
            print(
                "Существуют единичные корни, ряд не стационарен"
            )
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print()

    adf_check(df[column])
    df[column + "First Difference"] = df[column] - df[column].shift(1)
    df["Seasonal Difference"] = df[column] - df[column].shift(365)
    df["Seasonal First Difference"] = df[column + "First Difference"] - df[column + "First Difference"].shift(365)
    adf_check(df["Seasonal First Difference"].dropna())

    return df
```

```python
df = stationarity_check(df, 'temperatureMean')
```

```python
def acf_pacf_plot(df):
    acf_pacf_plot = plt.figure(figsize=(12, 6))
    ax1 = acf_pacf_plot.add_subplot(211)
    acf_pacf_plot = plot_acf(df['temperatureMean'].dropna(), lags=30, color="orange", ax=ax1)
    ax2 = acf_pacf_plot.add_subplot(212)
    acf_pacf_plot = plot_pacf(df['temperatureMean'].dropna(), lags=30, color="green", ax=ax2)
    plt.savefig("plots/acf_pacf_plot.png", dpi=200, figure=acf_pacf_plot)
```

```python
acf_pacf_plot(df)
```

```python
period = 365
```

```python
def model(df, period):

    model = sm.tsa.statespace.SARIMAX(
        df["temperatureMean"], order=(1, 1, 2), seasonal_order=(0, 1, 0, period)
    )
    results = model.fit()
    print(results.summary())
    return results
```

```python
%%time
results = model(df, period)
```

```python
def validate_model(df, results):
    
    """Я обнаружил, что SATIMAX от statsmodels
    в идеале не следует использовать для сезонных данных, которые не
    являются месячными, еженедельными. Промежуточный период = 365, 
    создает слишком много плотных массивов, и в результате компьютер,
    используемый для тестирования, исчерпывает все ОЗУ. На языке R 
    написаны пакеты, которые решают данную проблему"""
    
    forecast = results.predict(start=len(df), end=len(df)+70)
    df_test['SARIMA'] = forecast
    
    predict_existing_values_plot = plt.figure(figsize=(16, 8))
    ax = predict_existing_values_plot.add_subplot(111)
    plt.title(f"Forecast of Temperature", figure=predict_existing_values_plot)
    plt.xlabel("Time", figure=predict_existing_values_plot)
    plt.ylabel("Mean Temperature (˚C)", figure=predict_existing_values_plot)
    
    plt.plot(df["temperatureMean"].iloc[-500:], label='Train')
    plt.plot(df_test["temperatureMean"], label='Test')
    plt.plot(df_test['SARIMA'], label='SARIMA')
    
    plt.legend(loc='best')
    plt.show()
    
    predict_existing_values_plot.savefig(
        "plots/predict_plot.png", dpi=200)
```

```python
validate_model(df, results)
```
