### ÐœÐµÑ‚Ð¾Ð´ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð³Ð¾ Ñ‚Ñ€ÐµÐ½Ð´Ð° Ð¥Ð¾Ð»Ñ‚Ð°-Ð£Ð¸Ð½Ñ‚ÐµÑ€ÑÐ°

ÐŸÐ¾Ð´Ñ…Ð¾Ð´Ð¸Ñ‚ Ð´Ð»Ñ Ñ€Ð°Ð±Ð¾Ñ‚Ñ‹ Ñ Ð²Ñ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ð¼Ð¸ Ñ€ÑÐ´Ð°Ð¼Ð¸, Ð³Ð´Ðµ ÐµÑÑ‚ÑŒ ÑƒÑÑ‚Ð¾Ð¹Ñ‡Ð¸Ð²Ñ‹Ð¹ Ð»Ð¸Ð½ÐµÐ¹Ð½Ñ‹Ð¹ Ñ‚Ñ€ÐµÐ½Ð´.

Ð£ Ð½Ð°Ñ ÐµÑÑ‚ÑŒ ÑÐ»ÐµÐ´ÑƒÑŽÑ‰Ð¸Ðµ ÐºÐ¾Ð¼Ð¿Ð¾Ð½ÐµÐ½Ñ‚Ñ‹:

- Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ðµ ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ Ð´Ð»Ñ Ð¿Ð¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ñ Ð¿Ñ€Ð¾Ð³Ð½Ð¾Ð·Ð°    
${y}_{x+1} = \ell_x + b_x$    
- ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ ÑƒÑ€Ð¾Ð²Ð½Ñ    
$\ell_x = \alpha y_x + (1-\alpha)(\ell_{x-1} + b_{x-1})$
- ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ Ñ‚Ñ€ÐµÐ½Ð´Ð°    
${b_x} = \beta(\ell_x - \ell_{x-1}) + (1-\beta)b_{x-1}$


Ð˜Ñ‚Ð¾Ð³Ð¾Ð²Ð¾Ðµ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ðµ Ð¿Ñ€ÐµÐ´ÑÑ‚Ð°Ð²Ð»ÑÐµÑ‚ ÑÐ¾Ð±Ð¾Ð¹ ÑÑƒÐ¼Ð¼Ñƒ Ð¼Ð¾Ð´ÐµÐ»ÑŒÐ½Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ ÑƒÑ€Ð¾Ð²Ð½Ñ Ð¸ Ñ‚Ñ€ÐµÐ½Ð´Ð°.

```python
def double_exponential_smoothing(series, alpha, beta):
    result = [series[0]]
    for n in range(1, len(series)+1):
        if n == 1:
            level, trend = series[0], series[1] - series[0]
        if n >= len(series): # Ð¿Ñ€Ð¾Ð³Ð½Ð¾Ð·Ð¸Ñ€ÑƒÐµÐ¼
            value = result[-1]
        else:
            value = series[n]
        last_level, level = level, alpha*value + (1-alpha)*(level+trend)
        trend = beta*(level-last_level) + (1-beta)*trend
        result.append(level+trend)
    return result[-1]
```

```python
fit1 = Holt(np.asarray(train.orders)).fit(smoothing_level = 0.28,smoothing_slope = 0.27)
y_hat['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(16,8))
plt.plot(train.orders, label='Train')
plt.plot(test.orders, label='Test')
plt.plot(y_hat['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.orders, y_hat.Holt_linear))
print(rms)
```

### ÐœÐµÑ‚Ð¾Ð´ Ð¥Ð¾Ð»Ñ‚Ð°-Ð£Ð¸Ð½Ñ‚ÐµÑ€ÑÐ°

Ð¢ÐµÐ¿ÐµÑ€ÑŒ Ð±ÑƒÐ´ÐµÐ¼ ÑƒÑ‡Ð¸Ñ‚Ñ‹Ð²Ð°Ñ‚ÑŒ ÑÐµÐ·Ð¾Ð½Ð½Ð¾ÑÑ‚ÑŒ.

ÐŸÐ¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ Ð½Ð¾Ð²ÑƒÑŽ ÑÐ¸ÑÑ‚ÐµÐ¼Ñƒ:

$\ell_x = \alpha(y_x - s_{x-L}) + (1-\alpha)(\ell_{x-1} + b_{x-1})$


$b_x = \beta(\ell_x - \ell_{x-1}) + (1-\beta)b_{x-1}$

$s_x = \gamma(y_x - \ell_x) + (1-\gamma)s_{x-L}$

${y}_{x+m} = \ell_x + mb_x + s_{x-L+1+(m-1)modL}$


```python
fit1 = ExponentialSmoothing(np.asarray(train.orders) ,seasonal_periods=7 ,trend='add', seasonal='add').fit()
y_hat['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot(train.orders, label='Train')
plt.plot(test.orders, label='Test')
plt.plot(y_hat['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
rms = sqrt(mean_squared_error(test.orders, y_hat.Holt_Winter))
print(rms)
```

```python
import sys
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm

import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
from scipy.optimize import minimize

import matplotlib.pyplot as plt
```

```python
ps = range(1, 6)
ds = 1
qs = range(1, 6)
Ps = range(0, 4)
Ds = 1
Qs = range(0, 3)

from itertools import product

parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_rms = float("inf")
for t in range(40,60,5):
    train = restaurant_orders.sort_values('date', ascending = True).iloc[t:-18]

    for param in tqdm(parameters_list):
        #try except Ð½ÑƒÐ¶ÐµÐ½, Ð¿Ð¾Ñ‚Ð¾Ð¼Ñƒ Ñ‡Ñ‚Ð¾ Ð½Ð° Ð½ÐµÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ñ… Ð½Ð°Ð±Ð¾Ñ€Ð°Ñ… Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð² Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð½Ðµ Ð¾Ð±ÑƒÑ‡Ð°ÐµÑ‚ÑÑ
        try:
            model=sm.tsa.statespace.SARIMAX(train.orders.squeeze(), order=(param[0], ds, param[1]), 
                                        seasonal_order=(param[2], Ds, param[3], 7)).fit(disp=-1)
            
            forecast = model.predict(start = train.shape[0], end = train.shape[0]+18)
            y_hat['SARIMA'] = forecast
            rms = sqrt(mean_squared_error(test.orders, y_hat.SARIMA))

            #ÑÐ¾Ñ…Ñ€Ð°Ð½ÑÐµÐ¼ Ð»ÑƒÑ‡ÑˆÑƒÑŽ Ð¼Ð¾Ð´ÐµÐ»ÑŒ, aic, Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹
            if rms < best_rms:
                best_model = model
                best_rms = rms
                best_param = param
            results.append([t, param, rms])
            #Ð²Ñ‹Ð²Ð¾Ð´Ð¸Ð¼ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹, Ð½Ð° ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ñ… Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð½Ðµ Ð¾Ð±ÑƒÑ‡Ð°ÐµÑ‚ÑÑ Ð¸ Ð¿ÐµÑ€ÐµÑ…Ð¾Ð´Ð¸Ð¼ Ðº ÑÐ»ÐµÐ´ÑƒÑŽÑ‰ÐµÐ¼Ñƒ Ð½Ð°Ð±Ð¾Ñ€Ñƒ
        except Exception:
            print('wrong parameters:', param)
            continue

warnings.filterwarnings('default')
```
